1. Smooth the image `MACSJ0416_HAWKI.fits` with `scipy.ndimage.gaussian_filter`
using a kernel size of your preference (please explore!). Subtract the smoothed image
from the original one. Save the smoothed and subtracted images in fits format (using
astropy) and display them with with e.g. ginga (or ds9)

```python
from astropy.io import fits
from scipy.ndimage import gaussian_filter
import matplotlib.pyplot as plt
import numpy as np

hdu = fits.open("MACSJ0416_HAWKI.fits")
data = hdu[0].data
head = hdu[0].header

filtered = gaussian_filter(data, sigma=6)                 

subtracted = data - filtered

vmin = np.median(subtracted) - 1*np.std(subtracted)
vmax = vmin + 2*np.std(subtracted)

plt.figure(figsize=(10,10))
plt.imshow(subtracted, vmin=vmin, vmax=vmax, origin="lower", cmap="gray_r")

new_hdu = fits.ImageHDU(data=subtracted, header=head)
new_hdu.writeto("subtracted_image.fits")
```

The file `SDSS_spec_galaxy.csv` contains the spectrum of a star forming galaxy taken
as part of the Sloan Digital Sky Survey. Load the spectrum with astropy and make a
plot with matplotlib.
Knowing that the $H\alpha$ line at $\lambda = 6562.8$ Angstrom (restframe) is the most intense line
in this spectrum, please determine the redshift of this galaxy
Deredshift the spectrum and plot it over the original one

```python
tbl = Table.read("SDSS_spec_galaxy.csv")
wave = tbl["Wavelength"]
flux = tbl["Flux"]


plt.figure(figsize=(12, 4))
plt.plot(wave, flux)

peak = np.max(flux)
index = np.where(flux==peak)
wave_max = wave[index[0]]

redshift = (wave_max - 6562.8) / 6562.8

plt.plot(wave / (1 + redshift), flux)


plt.xlabel("wavelength")
plt.ylabel("flux")
```

Plot again the data table `rotcurve.txt` provided with this exercise. It contains data
for a rotation curve for a galaxy with columns `r` for radius, `vel` for velocity and the
error in vel. Load the data now with `astropy.table.Table`
W
hat is the maximum rotation velocity of this galaxy? Try to determine it by fitting
the following function to the data using `scipy.optimize.curve_fit`

$V (r) = 2\pi V_{max} \arctan(r/rd)$,

where `rd` is the disk scale length, ∼ 5 in this case, but can also be left as a free
parameter. Overplot the fit

```python
from scipy.optimize import curve_fit
tbl = Table.read("rotcurve.txt", format="ascii")
r = tbl["r"]
vel = tbl["vel"]
e_vel = tbl["e_vel"]

def func(r, vmax, rd):
    f = (2/np.pi)*vmax*np.arctan(r/rd)
    return f


p0 = np.array([100, 2])  # guesses
coeffs, matcov = curve_fit(func, r, vel, p0)

vmax = coeffs[0]
rd = coeffs[1]

plt.figure(figsize=(10,7))
plt.errorbar(r, vel, e_vel)
plt.plot(r, func(r, vmax, rd))
plt.xlabel("Distance [kpc]")
plt.ylabel("Velocity [km/s]")

print("Vmax: %s km/s, Rd: %s kpc" % (vmax, rd))

```
