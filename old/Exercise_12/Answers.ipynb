```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
import scipy.ndimage as ndimage
from astropy.io import fits
```

# Question 1

The fits file `MACSJ0416_HAWKI.fits` is a near infrared (_Ks_-band) image of the center of the
cluster `MACS J0416.1-2403` at $z=0.4$ taken with the HAWK-I instrument at the VLT as part
of the ESO contribution to the FrontierFields program (see http://gbrammer.github.io/HAWKI-FF/).

```python
hdu = fits.open("MACSJ0416_HAWKI.fits")
data= hdu[0].data
header = hdu[0].header

plt.figure(figsize=(12,12))
fig, ax = plt.subplots(figsize=(12,12))
plt.imshow(data, origin="lower", cmap="plasma", vmin=np.median(data), vmax=np.mean(data)+np.std(data))

maximum = np.max(data)
(y,x) = np.where(data == maximum)

circle = plt.Circle((x,y), 20, color='k', fill=False, lw=4)
ax.add_artist(circle)
```

```python
smooth = ndimage.gaussian_filter(data, 8)
fig, ax = plt.subplots(figsize=(12,12))
plt.imshow(smooth, origin="lower", cmap="plasma", vmin=np.median(data), vmax=np.mean(data)+np.std(data))
subtract = data - smooth
plt.imshow(subtract, origin="lower", cmap="plasma", vmin=np.median(data), vmax=np.mean(data)+np.std(data))
```
